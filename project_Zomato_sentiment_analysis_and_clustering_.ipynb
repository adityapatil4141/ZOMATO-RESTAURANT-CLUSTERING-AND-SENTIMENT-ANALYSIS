````python
# Required imports
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.stem.porter import PorterStemmer
import string
````

````python
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
````

````python
def text_process(text):
  tokens = nltk.word_tokenize(text)

  # Removing stopwords:
  stop_words= set(stopwords.words('english'))
  tokens = [word.lower() for word in tokens if (word.isalpha() and word.lower() not in stop_words and word not in string.punctuation)]

  # Lemmatization:
  lemmatizer = WordNetLemmatizer()
  tokens = [lemmatizer.lemmatize(word) for word in tokens]

  #Stemming:
  ps = PorterStemmer()
  tokens = [ps.stem(word) for word in tokens]

  text_processed = " ".join(tokens)
  return text_processed
````

````python
# Applying function to 'Review' column:
df_reviews['processed_reviews'] = df_reviews['Review'].apply(text_process)
````

````python
# Comparing old review column with processed review column:
df_reviews[['Review','processed_reviews']].head()
````

````python
def filter_word(text):
  tokens = word_tokenize(text)
  filtered_tokens = [word for word in tokens if 2<len(word)<45]
  return " ".join(filtered_tokens)
````

````python
from textblob import TextBlob
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
import plotly.express as px
````

````python
def subjectivity(text):
  return TextBlob(text).sentiment.subjectivity
````

````python
def polarity(text):
  return TextBlob(text).sentiment.polarity
````

````python
df_reviews['Subjectivity'] = df_reviews['processed_reviews'].apply(subjectivity)
df_reviews['Polarity'] = df_reviews['processed_reviews'].apply(polarity)
````

````python
def get_analysis(score):
  if score<0:
    return 'Negative'
  elif score == 0:
    return 'Neutral'
  else:
    return 'Positive'
````

````python
df_reviews.head()
````

````python
df_reviews['Analysis'] = df_reviews['Polarity'].apply(get_analysis)
````

````python
plt.figure(figsize=(10,6))
fig = px.scatter(df_reviews,x='Polarity',y = 'Subjectivity',color = 'Analysis',size='Subjectivity')
fig.show()
````

````python
# Converting analysis into numerical form:
def sentiment(analysis):
  if analysis=='Positive':
    return 0
  elif analysis =='Neutral':
    return 0
  else:
    return 1
````

````python
df_reviews['Analysis'] = df_reviews['Analysis'].apply(sentiment)
````

````python
sentiment_df = df_reviews[['processed_reviews','Rating','Analysis']]
````

````python
x= sentiment_df['processed_reviews']
y = sentiment_df['Analysis']
````

````python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,stratify=sentiment_df['Analysis'])
````

````python
# Creating tokens for text data:
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
x_train = vectorizer.fit_transform(x_train)
x_test = vectorizer.transform(x_test)
````

#Algorithm Implementaion

##MultinomialNB

````python
import sklearn.metrics as metrics
from sklearn.metrics import accuracy_score , confusion_matrix, roc_auc_score , classification_report , f1_score
````

````python
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()
````

````python
mnb.fit(x_train,y_train)
````

````python
y_pred_test = mnb.predict(x_test)
y_pred_train = mnb.predict(x_train)
````

````python
def model_details(y_train,y_test,y_pred_train,y_pred_test):
  print('-'*50)
  print("training accuracy = ",accuracy_score(y_train,y_pred_train))
  print('-'*50)
  print("test accuracy = ", accuracy_score(y_test,y_pred_test))
  print('-'*50)
  confusion_mat=confusion_matrix(y_pred_test,y_test)
  print("Confusion matrix on test data = ")
  print(confusion_mat)
  print('-'*50)
  classification_rep = classification_report(y_pred_test,y_test)
  print("classification_report =")
  print(classification_rep)
````

````python
model_details(y_train,y_test,y_pred_train,y_pred_test)
````

````python
 df_names.head()
````

````python
# Converting cuisines column values in lower.
df_names['Cuisines'] = df_names['Cuisines'].apply(lambda x : x.lower())
````

````python
# Taking necessary colums form names dataset and creating new dataframe cusisines_df
cuisines_df = df_names[['Restaurant','Cost','Cuisines']]
````

````python
cuisines_df.head()
````

````python
# Removing spaces form the vales of cuisines column and seperating them with ',' and converting it in list.
cuisines_df['Cuisines'] = cuisines_df['Cuisines'].str.replace(' ','')
cuisines_df['Cuisines'] = cuisines_df['Cuisines'].str.split(',')
````

````python
cuisines_df.head()
````

````python
# Importing multilabelbinarizer
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer(sparse_output=True)
````

````python
# Making one hot encoded labels for each type of cuisines and dropping the cuisines column.
cuisines_df = cuisines_df.join(pd.DataFrame.sparse.from_spmatrix(mlb.fit_transform(cuisines_df.pop('Cuisines')),
                                                                 index=cuisines_df.index, columns=mlb.classes_))
````

````python
cuisines_df.head()
````

````python
# Unique values in rating column
df_reviews['Rating'].unique()
````

````python
# Remove nan rating from rating column
df_reviews.dropna(subset=['Rating'],inplace=True)
````

````python
# change 'LIKE' rating to 4 in rating column as 4 was not in rating column:
df_reviews['Rating'] = df_reviews['Rating'].replace('Like','4')
````

````python
# Changing datatype of rating column:
df_reviews['Rating'] = df_reviews['Rating'].astype('float64')
````

````python
df_reviews = df_reviews.dropna(subset=['Rating'])
````

````python
# Creating new column to get average rating:
ratings_df = df_reviews.groupby('Restaurant')['Rating'].mean().reset_index()
````

````python
ratings_df.sort_values(by= 'Rating',ascending=False).head()
````

````python
# Merge
df_cluster = cuisines_df.merge(ratings_df,left_on='Restaurant',right_on='Restaurant')
````

````python
# Column names
df_cluster.columns
````

````python
# Changing names and order of columns.
df_cluster = df_cluster[['Restaurant','Cost','Rating','american','andhra','arabian','asian','bbq','bakery','beverages','biryani','burger','cafe','chinese','continental','desserts','european','fastfood','fingerfood','goan','healthyfood','hyderabadi','icecream','indonesian','italian','japanese','juices','kebab','lebanese']]
````

````python
# Dataset information.
df_cluster.info()
````

````python
# Cost can be in thousands (eg : 1,200) so we replace ',' with ''
df_cluster['Cost'] = df_cluster['Cost'].replace(',','')
````

````python
# Converting cost column in float datatype.
df_cluster['Cost'] = df_cluster['Cost'].astype('float64')
````

````python
# Plotting lmplot. lmplot are  scatter plots with overlaid regression lines
sns.lmplot(y='Rating',x='Cost',data=df_cluster,line_kws={'color':'red'},height=6.27,aspect=11.7/8.27)
````

##KMeans clustering

````python
# Necessary libraries imports
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from yellowbrick.cluster import KElbowVisualizer
````

````python
# Kmeans clustering:
# Elbow plot way1:
scores = [KMeans(n_clusters=i+2,random_state=11).fit(df_cluster.drop('Restaurant',axis=1)).inertia_ for i in range(8)]
plt.figure(figsize=(10,6))
sns.lineplot(x=np.arange(2,10),y=scores)
plt.xlabel("no. of clusters")
plt.ylabel("Inertia")
plt.show()
````

````python
# Way2
cluster_range = range(1, 8)
inertia_values=[]
for n_clusters in cluster_range:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(df_cluster.drop('Restaurant',axis=1))
    inertia_values.append(kmeans.inertia_)

# Plot the elbow plot
plt.plot(cluster_range, inertia_values, marker='o')
plt.title('Elbow Plot')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
````

````python
# Model implementation
model = KMeans(random_state=11,n_clusters=4)
model.fit(df_cluster.drop('Restaurant',axis=1))
````

````python
cluster_lbl = model.predict(df_cluster.drop('Restaurant',axis=1))
````

````python
df_cluster['label']=cluster_lbl
````

````python
plt.figure(figsize=(10,6))
sns.set_style('white')
sns.scatterplot(y='Rating',x='Cost',data=df_cluster, hue='label',palette='deep')
````

````python
cluster_0 = df_cluster[df_cluster['label']==0].reset_index()
cluster_1 = df_cluster[df_cluster['label']==1].reset_index()
cluster_2 = df_cluster[df_cluster['label']==2].reset_index()
cluster_3 = df_cluster[df_cluster['label']==3].reset_index()
````

````python
list_of_clusters = [cluster_0,cluster_1,cluster_2,cluster_3]
````

````python
for i,df in enumerate(list_of_clusters):
  print(f'Top cuisines in clusters {i}\n',df.drop(['index','Restaurant','Cost','Rating','label'],axis=1).sum().sort_values(ascending=False)[:3],'\n')
````
